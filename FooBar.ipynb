```
import math

def solution(area):
    # Initialize the result list.
    rst = []
    
    # Run it until the area left become zero.
    while area > 0:
        # Find the maximum square area smaller than current area.
        max_square_area = int(math.sqrt(area)) ** 2
        
        # Add max_square_area as many as possible to the result list.
        while area >= max_square_area:
            rst.append(max_square_area)
            area -= max_square_area
            
    return rst
```
